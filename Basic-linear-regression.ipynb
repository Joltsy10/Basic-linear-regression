```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('data.csv')
x = df.iloc[:,0].values
y = df.iloc[:,-1].values
mymodel = []
l = 0.00001
```

```python
w = 0
b = 0
```

```python
def costfunc():
    predictions = w * x + b
    return np.sum((predictions - y) ** 2) / (2 * len(x))

def gradient_descent():
    global w,b
    prediction = w * x + b
    error = prediction - y
    temp_w = np.dot(error,x)/len(x)
    temp_b = np.sum(error)/len(x)
    w = w - l * temp_w
    b = b - l * temp_b
```

```python
for n in range(10):
    gradient_descent()
costfunc()
```

```python
mymodel = w * x + b
```

```python
plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show
```
