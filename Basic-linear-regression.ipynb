```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('data.csv')
x = df.iloc[:,:1].values
y = df.iloc[:,-1].values
mymodel = []
l = 0.0003
```

```python
w = 0
b = 0
```

```python
def func(a):
    return w*a + b

def sigma2():
    total = 0
    for n in range(len(x)):
        total += (func(x[n]) - y[n])*x[n]
    return total

def sigma3():
    total = 0
    for n in range(len(x)):
        total += func(x[n]) - y[n]
    return total

def costfunc():
    predictions = w * x + b
    return np.sum((predictions - y) ** 2) / (2 * len(x))

def gradient_descent():
    global w
    global b
    temp_w = w - l/len(x)*sigma2()
    temp_b = b - l/len(x)*sigma3()
    w = temp_w
    b = temp_b
```

```python
for n in range(1000):
    gradient_descent()
costfunc()
```

```python
mymodel = w * x + b
```

```python
plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show
```
