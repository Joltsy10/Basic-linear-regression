```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('data.csv')
x = df.iloc[:,0].values
y = df.iloc[:,-1].values
mymodel = []
l = 0.001
```

```python
x_norm = (x - np.mean(x)) / np.std(x)
y_norm = (y - np.mean(y)) / np.std(y)

```

```python
w = 0
b = 0
```

```python
def costfunc():
    predictions = w * x_norm + b 
    return np.sum((predictions - y_norm) ** 2) / (2 * len(x))

def gradient_descent():
    global w,b
    prediction = w * x_norm + b
    error = prediction - y_norm
    temp_w = np.dot(error,x_norm)/len(x)
    temp_b = np.sum(error)/len(x)
    w = w - l * temp_w
    b = b - l * temp_b
```

```python
for n in range(1000):
    gradient_descent()
costfunc()
```

```python
mymodel = w * x_norm + b
mymodel_original = mymodel * np.std(y) + np.mean(y)
```

```python
plt.scatter(x,y)
plt.plot(x,mymodel_original)
plt.show()
```
